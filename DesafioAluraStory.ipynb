### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


Confira como estão os dados de cada loja:

```python
# Visualizar as primeiras linhas de cada DataFrame
print(loja.head())
print(loja2.head())
print(loja3.head())
print(loja4.head())
```

Somar a coluna "Preço" de cada DataFrame para obter o faturamento total de cada loja:

```python
# Calculando o faturamento total de cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Exibindo os resultados
print('Faturamento Loja 1:', faturamento_loja1)
print('Faturamento Loja 2:', faturamento_loja2)
print('Faturamento Loja 3:', faturamento_loja3)
print('Faturamento Loja 4:', faturamento_loja4)

```

Grafico:

```python
import matplotlib.pyplot as plt

# Dados de faturamento das lojas
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [1534509.12, 1488459.06, 1464025.03, 1384497.58]

plt.figure(figsize=(8, 8))
plt.pie(faturamentos, labels=lojas, autopct='%1.1f%%', startangle=90, colors=['#4F81BD','#C0504D','#9BBB59','#8064A2'])
plt.title('Participação no Faturamento Total por Loja')
plt.axis('equal')  # Deixa o gráfico redondo

plt.show()
```

# 2. Vendas por Categoria


Descobrir as categorias de produtos mais vendidas em cada loja, contar quantas vezes cada categoria aparece na coluna "Categoria do Produto" . Assim, você saberá quais categorias têm mais vendas.

```python
# Contando as categorias mais vendidas em cada loja
categorias_loja1 = loja['Categoria do Produto'].value_counts()
categorias_loja2 = loja2['Categoria do Produto'].value_counts()
categorias_loja3 = loja3['Categoria do Produto'].value_counts()
categorias_loja4 = loja4['Categoria do Produto'].value_counts()

print("Categorias mais vendidas - Loja 1:")
print(categorias_loja1)
print("\nCategorias mais vendidas - Loja 2:")
print(categorias_loja2)
print("\nCategorias mais vendidas - Loja 3:")
print(categorias_loja3)
print("\nCategorias mais vendidas - Loja 4:")
print(categorias_loja4)
```

# 3. Média de Avaliação das Lojas

Calculando a média de avaliação dos clientes das lojas:

```python
# Calculando a média de avaliação das lojas
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

# Exibindo os resultados
print(f'Média de Avaliação - Loja 1: {media_avaliacao_loja1:.2f}')
print(f'Média de Avaliação - Loja 2: {media_avaliacao_loja2:.2f}')
print(f'Média de Avaliação - Loja 3: {media_avaliacao_loja3:.2f}')
print(f'Média de Avaliação - Loja 4: {media_avaliacao_loja4:.2f}')
```

# 5. Frete Médio por Loja


```python
# Calculando o frete médio de cada loja
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Exibindo os resultados
print(f'Frete médio - Loja 1: R$ {frete_medio_loja1:.2f}')
print(f'Frete médio - Loja 2: R$ {frete_medio_loja2:.2f}')
print(f'Frete médio - Loja 3: R$ {frete_medio_loja3:.2f}')
print(f'Frete médio - Loja 4: R$ {frete_medio_loja4:.2f}')
```

```python
import matplotlib.pyplot as plt

# Dados do frete médio das lojas
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes = [34.69, 33.62, 33.07, 31.28]

plt.figure(figsize=(8,5))
barras = plt.bar(lojas, fretes, color='cornflowerblue')
plt.title('Frete Médio por Loja')
plt.xlabel('Loja')
plt.ylabel('Frete Médio (R$)')

# Adicionando os valores em cima de cada barra
for i, barra in enumerate(barras):
    y = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2, y + 0.2, f'R$ {y:.2f}',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
